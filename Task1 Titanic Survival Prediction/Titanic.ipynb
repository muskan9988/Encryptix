```python
import numpy as np
import pandas as pd
import sklearn
```

```python
data = pd.read_csv('Titanic-Dataset.csv')
# data = pd.DataFrame(data)
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Cabin'] = data['Cabin'].fillna(data['Cabin'].mode()[0])
data.isnull().sum()
```

```python
data.head(5)
```

<h3>All tha Data has  been PreProcessed we have remove the null values and dropped some categories like Embarked and Cabin which ha ve no significance in analysis<h3>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Sex']  =le.fit_transform(data['Sex'])   # male =1 , female = 0
data['Embarked'] = le.fit_transform(data['Embarked'])
data
```

```python
data = data.drop(columns=['Name','Ticket','Cabin']) # droping the columns with string data type 
data
```

```python
data.corr()
```

```python
import seaborn as sns
sns.heatmap(data.corr(),annot=True)
```

From the heatmap we can visualise that Survived parameter is strongly dependent on <b style="color:green">Pclass, Fare, Sex</b> so we will use this feature to predict the survival of the passengers in the model

```python
X = data[['PassengerId','Pclass','Fare','Sex']]
y = data['Survived']
print(type(X),type(y))
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_X_train = scaler.fit_transform(X_train)
sc_X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
model = SVC(kernel='linear',gamma='auto',random_state=42)
# model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict on the validation set
y_pred = model.predict(X_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
print(confusion_matrix(y_test, y_pred))
```
